```python
data=data.iloc[1:]
from sklearn import svm
clf = svm.SVC(gamma='scale')
x=data.drop('status',axis=1)
y=data['status']


#SPLITING the dataset into test and training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)

#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
#make the ANN model
#Import the Keras
#import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense

#initialising the ANN
classifier = Sequential()

#Adding the first inpit layer and hidden layer
classifier.add(Dense(output_dim = 11, init = 'uniform', activation = 'relu',input_dim = 20))

#Adding Second hidden layer
classifier.add(Dense(output_dim = 11, init = 'uniform', activation = 'relu'))

#adding output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

##fitting ANN to training Set
classifier.fit(X_train,y_train, batch_size = 10, nb_epoch = 100)

y_pred = classifier.predict(X_test)
```
